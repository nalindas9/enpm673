```python
import numpy as np
import cv2
import copy
#import imutils
import matplotlib.pyplot as plt

   
def contour_gen2(frame):
    im = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    blur_img = cv2.GaussianBlur(im, (5,5), 0)
    edge = cv2.Canny(blur_img,100,200)
    edge1 = copy.copy(edge)
    contours,hierarchy = cv2.findContours(edge1, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    hierarchy = hierarchy[0] #innermost hierarchy
    
    corner_list = list()
     
    contour_list = list()
    
    for component in zip(contours,hierarchy):
        currentContour = component[0]
        currentHierarchy = component[1]
        perimeter = cv2.arcLength(currentContour, True)
        contourCorner = cv2.approxPolyDP(currentContour,0.015*perimeter,True)
        if len(contourCorner)==4 and currentHierarchy[3]!=-1: # and cv2.contourArea(contourCorner)<7000:
            #contourCorner=contourCorner.reshape(-1,2)
            #print("Contour Approximator:",contourCorner)
            #print("1",contourCorner[0][0],"2",contourCorner[1][0],"3",contourCorner[2][0],"4",contourCorner[3][0])
            corner_list.append(contourCorner)
            
    for corners in corner_list:
        area = cv2.contourArea(corners)
        print("Area contour :",area)
        if area>100 and area<7000:
            contour_list.append(corners)

    print("Length of contour:",len(contour_list))
    
    cont_img = cv2.drawContours(frame, contour_list, -1, (0,255,0),2)    
    cv2.imshow('AR Tag - Detected',cont_img)
```

```python
cap = cv2.VideoCapture('/home/akhopkar/Downloads/Tags1.mp4')
#Check condition for multiple tags

if cap.isOpened() == False:
    print("Error loading")
Frame = 0
while cap.isOpened():
    ret,frame = cap.read()
    print("Frame:",Frame)
    if ret == False:
         break
    img = cv2.resize(frame, (0,0),fx = 0.5, fy = 0.5)
    #contour_generate_(img)
    contour_gen2(img)
    Frame+=1
    if cv2.waitKey(5) & 0xFF == 27:
         break
            
cap.release()
cv2.destroyAllWindows()
```

```python
def contour_generate_(frame):
    im = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    blur_img = cv2.GaussianBlur(im, (5,5), 0)
    edge = cv2.Canny(blur_img,100,200)
    edge1 = copy.copy(edge)
    contours,hierarchy = cv2.findContours(edge1, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    #hierarchy = hierarchy[0] #innermost contours
    hierarchy = hierarchy[0]
    
    contours_ = list()
    corner_list = list()
    ext_index =list() 
    contour_list = list()
            
    for h in hierarchy:
        if h[3] != -1:
            ext_index.append(h[3]) #index of all the parent
            #ext_index.append(h[2])
    
    #Looping over the exterior contours
    for i in ext_index:
        perimeter = cv2.arcLength(contours[i],True)
        approx = cv2.approxPolyDP(contours[i],0.01*perimeter,True)
        
        if len(approx) == 4: #and cv2.contourArea(approx) < 7000: #If more than 4 corners detected, take the previous values into consideration
            #peri = cv2.arcLength(contours[i-1],True)
            #corners = cv2.approxPolyDP(contours[i-1], 0.02*peri, True)
            #corner_list.append(corners)
            contour_list.append(approx)
    #print("Length of corners:",len(corner_list))
    #Append valid corners into contour list
#     for corner in corner_list:
#         if len(corner) == 4:
#             contour_list.append(corner)
    print("contour_list:",len(contour_list))
    if len(contour_list)>0:
        for contour in contour_list:
            area = cv2.contourArea(contour)
            print("Area = ",area)

            #The following needs tuning
            if area > 300 and area < 7000: #Need to tune these
            #if area < 7000:
                #print("length contours: ",len(contours))
                contours_.append(contour)

#     print("Contours ",contours_)
    cont_img = cv2.drawContours(frame, contours_, -1, (0,255,0),2)    
    cv2.imshow('AR Tag - Detected',cont_img)
    print("Final contours ",contours_)
    print("Length of final contour",len(contours_))
    return contours_
```
